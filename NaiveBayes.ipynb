```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

```python
dataset = pd.read_csv(r"E:\aiml\dataset.csv")
dataset.head(10)
df = dataset.copy()
df.head()
```

```python
columns_to_encode = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']
label_encoder = LabelEncoder()
for column in columns_to_encode:
  df[column] = label_encoder.fit_transform(df[column])
df.head()
```

```python
X = df.iloc[:,:-1].values
Y = df.iloc[:,-1].values.reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.3, random_state = 42)
```

```python
gnb = GaussianNB()
y_pred = gnb.fit(x_train, y_train).predict(x_test)
```

```python
columns = dataset.iloc[:,:-1].columns

columns = columns.tolist()

columns.extend(["Actual Values", "Predicted Values"])
result = pd.DataFrame(data=np.c_[x_test[10:20],y_test[10:20],y_pred[10:20]], columns= columns)
result
```
